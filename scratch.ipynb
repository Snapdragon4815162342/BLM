```python
import pandas as pd
```

```python
uniprot_reference = pd.read_csv(r"C:\Users\lorenzo\Desktop\BLM\bacteria\uniprot_reference_proteomes_ids.csv", sep = ';')
uniprot_reference
```

```python
embl_meta = pd.read_csv(r"C:\Users\lorenzo\Desktop\datasets\phagescope\Metadata\phage_metadata\embl_phage_meta_data.tsv", sep = '\t')
embl_meta
```

```python
# Split the "Host" column into multiple rows
bacteria_series = embl_meta['Host'].str.split(';').explode()

# Remove any empty strings that might result from splitting
bacteria_series = bacteria_series[bacteria_series != '']
bacteria_series
```

```python
# Clean bacterial names
def clean_bacteria_name(name):
    name = name.replace('[', '').replace(']', '').strip()  # Remove brackets
    if 'sp.' in name:  # Handle species abbreviations
        name = name.split('sp.')[0].strip()
    return name

bacteria_series = bacteria_series.apply(clean_bacteria_name).drop_duplicates()
bacteria_series
#bacteria_series_cleaned
```

```python
uniprot_reference['Species Name']
```

```python
# Filter the DataFrame for bacteria
bacteria_df = uniprot_reference[uniprot_reference["SUPERREGNUM"] == "bacteria"]

# Create a function for loose matching
def find_proteome_id(name, species_column, proteome_column):
    for index, species in enumerate(species_column):
        if name in species:  # Check if the name is a substring of the species
            return proteome_column.iloc[index]
    return None

# Map the bacteria names in the Series to their Proteome_ID using loose matching
proteome_ids = bacteria_series.apply(
    lambda name: find_proteome_id(name, bacteria_df["Species Name"], bacteria_df["Proteome_ID"])
)

# Drop NaN values and output the result
proteome_ids = proteome_ids.dropna()
print(len(proteome_ids))  # Count of matches
print((proteome_ids))  # Matched Proteome_IDs
```

Select the proteins listed in n_only_surface.gff3 and create n_only surface.fasta

```python
import os
from Bio import SeqIO

# Folder containing the files
folder_path = "C:/Users/lorenzo/Desktop/Nuova cartella"

# Iterate through numbers corresponding to the files
for n in range(1, 1902):
    fasta_file = os.path.join(folder_path, f"{n}.fasta")
    gff3_file = os.path.join(folder_path, f"{n}_only_surface.gff3")
    output_file = os.path.join(folder_path, f"{n}_only_surface.fasta")

    # Skip if files don't exist
    if not os.path.exists(fasta_file) or not os.path.exists(gff3_file):
        print(f"Skipping {n}: Missing files.")
    

    # Extract headers from the gff3 file
    with open(gff3_file, 'r') as gff3:
        headers = [
            line.split(" ")[1]
            for line in gff3
            if line.strip() and line.startswith("#")
        ]
#print(headers)


    # Read sequences from the FASTA file and filter by headers
    with open(fasta_file, 'r') as fasta, open(output_file, 'w') as out_fasta:
        for record in SeqIO.parse(fasta, "fasta"):
            if record.id in headers:
                SeqIO.write(record, out_fasta, "fasta")

    print(f"Created {output_file}")
```
